#Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#Load the image in a variable in grayscale format

```python
path = '/content/drive/MyDrive/ImLab1/lena.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

#Declaring the kernel(Filter)

```python
kernel = np.array(([0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]), np.float32)
```

#Adding Border in our original image

```python
mimg = cv2.copyMakeBorder(img, 1,1,1,1, cv2.BORDER_REPLICATE)
```

#Taking an empty array for the resulting image

```python
result = np.zeros((img.shape[0],img.shape[1]), dtype="float32")
```

#Determining the kernel size range assuming the center is 0,0

```python
v = (3-1)//2
```

#Convolution

```python
for i in range(v, mimg.shape[0]-v):
  for j in range(v, mimg.shape[1]-v):
    sum = 0;
    for p in range(-v, v+1):
      for q in range(-v, v+1):
        sum = sum + kernel[p+v][q+v]*mimg[i-p][j-q]
    result[i-v][j-v] = sum/255
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
