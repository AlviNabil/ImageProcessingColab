```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp_img1 = cv2.imread("../img/apple.jpg",1)
inp_img2 = cv2.imread("../img/orange.jpg",1)
half_img1 = np.zeros_like(inp_img1)
half_img2 = np.zeros_like(inp_img2)
```
