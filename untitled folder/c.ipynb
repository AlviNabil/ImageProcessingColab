```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def _kernel(size=11):
    kernel = np.zeros((size, size), np.float32)
    for i in range(size):
        for j in range(size):
            if i==j:
                kernel[i,j]=1
    return kernel/np.sum(kernel)
```

```python
path = "../img/loki.jpg"
inp_img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
img= cv2.resize(inp_img, (400, 400))

plt.imshow(img, "gray")
plt.show()
```

```python
kernel = _kernel(11)
plt.imshow(kernel)
```

```python
padding = (img.shape[0]//2-kernel.shape[0]//2)
padded_kernel = np.pad(kernel, (padding, padding-1), 'constant', constant_values=(0))
plt.imshow(padded_kernel)

```

```python

IMG = np.fft.fft2(img)
IMGs = np.fft.fftshift(np.log1p(np.abs(IMG)))

KERNEL = np.fft.fft2(padded_kernel)
KERNELs = np.fft.fftshift(np.log1p(np.abs(KERNEL)))
figure, axis = plt.subplots(1,2, figsize=(14, 6))
axis[0].imshow(np.abs(IMGs), "gray")
axis[1].imshow((np.abs(KERNELs)), "gray")
plt.show()
```

```python
Blurred_IMG = np.multiply(IMG,KERNEL)
B_IMG =np.fft.ifftshift( np.real(np.fft.ifft2(Blurred_IMG)))
plt.imshow(B_IMG,"gray")
```

```python
output = np.divide(Blurred_IMG, KERNEL)
output = np.real(np.fft.ifft2(output))
plt.imshow(output,"gray")
```
