```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "../img/labtest.png"
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img,"gray")
```

```python
t,img = cv2.threshold(img, 170, 255, cv2.THRESH_BINARY)

plt.imshow(img,"gray")
```

```python
se = np.ones((7,7), np.uint8)
se=se-1
se[3,0]=se[0,3]=se[6,3]=se[3,6]=se[1,2]=se[2,1]=se[1,4]=se[4,1]=se[2,5]=se[5,2]=se[4,5]=se[5,4]=1
se[3,3]=0
plt.imshow(se)
```

```python
result = cv2.erode(img, se,1)
plt.imshow(result,"gray")
cv2.imshow("img", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result = cv2.dilate(img, se,1)
plt.imshow(result,"gray")
```
