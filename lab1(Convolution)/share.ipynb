#Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#Load the image in a variable in grayscale format

```python
path = '../img/lena.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, "gray")
plt.show()
```

#Declaring the kernel(Filter)

```python
kernel = np.array(([0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]), np.float32)
```

#Adding Border in our original image

```python
mimg = cv2.copyMakeBorder(img, 1,1,1,1, cv2.BORDER_REPLICATE)
```

#Taking an empty array for the resulting image

```python
result = np.zeros((img.shape), dtype="float32")
```

#Determining the kernel size range assuming the center is 0,0

```python
v = (3-1)//2
```

#Convolution

```python
a,b = img.shape
m,n = kernel.shape
m=m//2
n=n//2
```

```python
output = np.zeros((mimg.shape), np.float32)
for i in range(a):
  for j in range(b):
    for p in range(-m,m+1):
      for q in range(-n, n+1):
        if i-p>=0 and i-p<a and j-q>=0 and j-q<b:
          output[i,j] += mimg[i-p, j-q]*kernel[p+m][q+n]
plt.imshow(output, "gray")
final_output = np.zeros((img.shape), np.float32)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    final_output[i,j] = output[i,j]
plt.imshow(final_output, "gray")
print(final_output.shape, img.shape)
```
